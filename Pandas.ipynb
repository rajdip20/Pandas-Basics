```python
def batsman_v_team(batsman_name):
    mask=delivery['batsman']==batsman_name
    batsman=delivery[mask]
    return batsman.groupby('bowling_team').sum()['batsman_runs'].sort_values(ascending=False).head(1).index[0]
```

```python
# Merge Dataframes

total=df.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_df=total[total['batsman']=='V Kohli']
```

```python
import matplotlib.pyplot as plt
batsman_df.groupby('season').sum()['batsman_runs'].plot(kind='bar')
```

```python
def career_graph(batsman_name1,batsman_name2):
    batsman_df1=total[total['batsman']==batsman_name1]
    batsman_df1.groupby('season').sum()['batsman_runs'].plot()
    batsman_df2=total[total['batsman']==batsman_name2]
    batsman_df2.groupby('season').sum()['batsman_runs'].plot()
```

```python
career_graph('DA Warner','V Kohli')
```

```python
titanic['family_type']="Nothing"
```

```python
def family_label(fsize):
    if fsize==0:
        return "Alone"
    elif fsize>0 and fsize<=3:
        return "Small"
    elif fsize>3 and fsize<=6:
        return "Medium"
    else:
        return "Large"
```

```python
family_label(2)
```

```python
titanic['family_type']=titanic['Family_size'].apply(family_label)
```
